# 1. Polar Coordinates
You are given a complex . Your task is to convert it to polar coordinates.

```python
import cmath
z = complex(input())
print("Absolute value:",abs(z))
print("Phase:",cmath.phase(z))
```

# 2. Find Angle MBC
image.png

```python
from math import atan, degrees
AB = int(input())
BC = int(input())
angleC = atan(AB/BC)

# Degree symbol is chr(176)
# Using thale's theorem in triangle BMC angle B = angle C
print("angle MBC :",round(degrees(angleC)),chr(176),sep='')
```

# 3. Triangle Quest 2
You are given a positive integer N.
Your task is to print a palindromic triangle of size N.

```python
# 1 , 11*11 , 111*111 .......
for i in range(1,int(input())+1): 
    print((10**i//9)**2)
```

# 4. Mod Divmod
Read in two integers,a  and b, and print three lines.
The first line is the integer division a//b (While using Python2 remember to import division from __future__).
The second line is the result of the modulo operator: a%b .
The third line prints the divmod of a and b.

```python
a = int(input())
b = int(input())

print("Divison:",a // b)
print("Modulus:",a % b)
print("Divmod",divmod(a, b))
```

# 5. Power - Mod Power
You are given three integers:a , b, and m . Print two lines.
On the first line, print the result of pow(a,b). On the second line, print the result of pow(a,b,m).

```python
a = int(input())
b = int(input())
m = int(input())

print("a^b:",pow(a,b))
print("a^b mod m:",pow(a,b,m))
```

# 6. Integers Come In All Sizes

As we know, the result of a^b grows really fast with increasing b.

Let's do some calculations on very large integers.

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())
print("Answer:",(a**b)+(c**d))
```
